```python
import numpy
from matplotlib import pyplot
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

```python
nx = 41 # try changing this number from 41 to 81 and Run All ... what happens? 'nx' is the number of spatial grid pts
dx = 2/(nx-1) # 'dx' is the distance between any pair of adjacent grid points
nt = 25 # 'nt' is the number of time steps
dt = .02 # time step
c = 1 # wave speed
x = numpy.linspace(0,2,nx)
```

```python
u = numpy.ones(nx) # numpy array of ones that is nx elements long
lbound = numpy.where(x >= 0.5)
ubound = numpy.where(x <= 1)

print(lbound)
print(ubound)
```

```python
bounds = numpy.intersect1d(lbound,ubound)
u[bounds] = 2 # setting u = 2 between 0.5 and 1 as per out initial conditions
print(u)
```

```python
pyplot.plot(x, u, color='#003366', ls='--', lw=3)
pyplot.ylim(0,2.5);
```

```python
for n in range(1,nt):
    un = u.copy()
    for i in range(1,nx):
        
        u[i] = un[i] - c*dt/dx*(un[i]-un[i-1])
```

```python
pyplot.plot(x, u, color='#003366', ls='--', lw=3)
pyplot.ylim(0,2.5);
# increasing the number of grid points gets the plots closer to the desired shape
```

```python
## problem parameters
nx = 41
dx = 2/(nx-1)
nt = 10
dt = .02

## initial conditions
u = numpy.ones(nx)
u[numpy.intersect1d(lbound, ubound)] = 2
```

```python
pyplot.plot(x, u, color='#003366', ls='--', lw=3)
pyplot.ylim(0,2.5);
```

```python
for n in range(1,nt):
    un = u.copy()
    u[1:] = un[1:]-un[1:]*dt/dx*(un[1:]-un[0:-1])
    u[0] = 1.0
```

```python
pyplot.plot(x, u, color='#003366', ls='--', lw=3)
pyplot.ylim(0,2.5);
# not symetrical anymore not because of error, but physics
```
